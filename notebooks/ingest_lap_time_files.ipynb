```
lap_time_schema = StructType([StructField("raceId", IntegerType(), False),
                             StructField("driverId", IntegerType(), False),
                             StructField("lap", IntegerType(), False),
                             StructField("position", IntegerType(), True),
                             StructField("time", StringType(), True),
                             StructField("milliseconds", IntegerType(), True)
                             ])
```

```
lap_time_df = spark.read \
.schema(lap_time_schema) \
.csv(f"{raw_folder_path}/lap_times/lap_times_split*.csv")
```

```
final_df = lap_time_df.withColumn("ingestion_date", current_timestamp()) \
.withColumnRenamed("raceId","race_id") \
.withColumnRenamed("driverId","driver_id")
```

```
final_df.write.mode("overwrite").parquet(f"{processed_folder_path}/lap_times")
```
